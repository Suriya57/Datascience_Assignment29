```python
# Importing modules
import numpy as np
from sklearn import cluster, datasets 
from scipy import misc
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Visualizng racoon image

c=misc.face(gray=True)
plt.imshow(c)
```

```python
# Reshaping the image to apply cluster model

racoon=c.reshape(c.shape[0]*c.shape[1],1)
```

```python
# Executing k means clustering with 6 clusters 

k_means = KMeans(n_clusters=5, random_state=0) 
clustr=k_means.fit(racoon)
clustr
```

```python
# Array representation of a cluster

clusters = np.asarray(k_means.cluster_centers_,dtype=np.uint8) 
clusters
```

```python
# Display the labels of the picture

labels = np.asarray(k_means.labels_,dtype=np.uint8 )  
labels
```

```python
labels.shape
```

```python
# Reshape the labels # Reshap 

labels=labels.reshape(c.shape[0],c.shape[1])
labels
```

```python
# Save the compressed Raccoon.

plt.imsave('compressed_racoon.png',labels);
```

```python
# Show the compressed image

image = plt.imread('compressed_racoon.png')
plt.imshow(image)
plt.show()
```

```python
# Compression comparision

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(c)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(image)
plt.show()
```
